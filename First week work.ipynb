```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np 

dates = pd.date_range('2025-01-01', periods = 90)
temps = pd.Series(15+10*np.sin(np.linspace(0, 3*np.pi,90)), index = dates)
df = pd.DataFrame({'temp':temps})

df['temp'].plot()
df['temp'].rolling(7).max().plot(style='g:')
plt.show()
```

```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np 

dates = pd.date_range('2025-01-01', periods = 90)
temps = pd.Series(15+10*np.sin(np.linspace(0, 3*np.pi,90)), index = dates)
df = pd.DataFrame({'temp':temps})

plt.plot(df.index, df['temp'], 'b-')
plt.plot(df.index, df['temp'].rolling(7).max(), linestyle=':', color='green')
plt.show()
```

```python
import pandas as pd
from pandas.api.types import CategoricalDtype
data = {
    'Feeling happy?': pd.Categorical(
        ['Strongly agree', 'Disagree', 'Agree', 'Indifferent', 'Indifferent', 'Agree', 'Strongly disagree', 'Strongly agree', 'Indifferent', 'Strongly disagree'],
        categories=['Strongly disagree', 'Disagree', 'Indifferent', 'Agree', 'Strongly agree'],
        ordered=True
    ),
    'Social class': pd.Categorical(
        ['II', 'III', 'II', 'I', 'II', 'I', 'IV', 'III', 'V', 'III'],
        categories=['I', 'II', 'III', 'IV', 'V'],
        ordered=True
    ),
    'Age': [25, 35, 44, 28, 35, 42, 36, 38, 30, 40]
}
df = pd.DataFrame(data)

ordinal_columns = [
    col for col in df.columns
    if isinstance(df[col].dtype, CategoricalDtype) and df[col].cat.ordered
]

print("Ordinal Variables:", ordinal_columns)
```

```python
import pandas as pd

df = pd.read_csv("Cars.csv")
df_clean = df.dropna(subset=["horsepower"])

max_hp = df_clean["horsepower"].max()
max_hp_cars = df_clean[df_clean["horsepower"] == max_hp][["name", "horsepower"]]
print("1. Car(s) with the highest horsepower:")
print(max_hp_cars)
print()
```

```python
import pandas as pd

df = pd.read_csv("Cars.csv")
high_mpg_cars = df[df["mpg"] >= 35]
print("2. Number of cars with mpg  35:", high_mpg_cars.shape[0])
print()
```

```python
import pandas as pd

df = pd.read_csv("Cars.csv")
mean_acc_japan = round(df[df["origin"] == "japan"]["acceleration"].mean(), 2)
print("4. Mean acceleration of cars from Japan:", mean_acc_japan)
print()
```

```python
import pandas as pd

df = pd.read_csv("Cars.csv")
avg_mpg_by_year = df.groupby("model_year")["mpg"].mean()
year_with_highest_avg_mpg = avg_mpg_by_year.idxmax()
highest_avg_mpg = round(avg_mpg_by_year.max(), 2)
print("5. Year with highest average mpg:", year_with_highest_avg_mpg)
print("   (Average mpg in that year:", highest_avg_mpg, ")")
```
